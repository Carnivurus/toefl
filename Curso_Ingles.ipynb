Donde puedo encontrar a los sustantivos(nouns) ?

- `1.- Al inicio de una oración (usualmente conocido como sujeto)`
- `2.- Despues de un verbo (se le conoce como objeto)`
- `3.- Después de una preposición (se le conoce como objeto)`
- `4.- Despues de un articulo (a, an, the) como sujeto`

## 3.- Uso del article 'The' con Nouns

- `'The'`puede ser usado con singulares, plulares e incontables.
- `'The'` puede ser omitido para plurales. Ejemplo (Cars are awesome)

El uso de `'The'` en singulares es para situaciones : `Unicas, especiales, detalladas que cuenten con contexto`

Uniques
- The president (único)
- The sun (único)
- The earth

Plurales
- Cars are fast (cars in general)
- The Cars are fast

Singulares* 
Generalmente deben venir acompañados (the apple, an apple)

## 4.- Adjetivos

Elements to describe an animal, person, thing

- The `fat` cat is here again! 
- `Those` times will never come back
- My `gentle` cousin lends me a towel
- I watched a very interesting movie yesterday - Algunos adjetivos tienen terminación ing, ed


`Antes del sustantivo (noun)` (atributivo): Esta es la posición más común para los adjetivos en inglés. Los adjetivos se colocan antes del sustantivo que modifican. Por ejemplo:

- "The red car" (El coche rojo)
- "A beautiful garden" (Un jardín hermoso)
- "His old house" (Su casa vieja)

`Después del verbo "to be" (predicativo)`: Los adjetivos también pueden colocarse después del verbo "to be" para describir al sujeto. En este caso, el adjetivo está predicando algo sobre el sujeto y no modifica directamente un sustantivo. Por ejemplo:

- "The car is red" (El coche es rojo)
- "The garden looks beautiful" (El jardín se ve hermoso)
- "His house seems old" (Su casa parece vieja)

Tipos de adjetivos: https://www.britishcouncil.org.mx/blog/adjetivos-ingles

- `El adjetivo suele ir antes del sustantivo`
- `Algunos verbos obligan a ubicar el adjetivo después del verbo` (to be, become, fell, look, to seem etc.) The child looks happy / Todos son estados o sentidos.
- `Se pueden usar varios adjetivos juntos` - A big red moon.

 `Si un adjetivo intenta describir algo que no sea un sustantivo, ese sería el error.`

 `Si hay un adjetivo pero no un noun, esta mal`


 Los adjetivos pueden superlativos o comparativos.


| Tipo de adjetivo | Comparativo | Superlativo |
|-------------------|-------------|-------------|
| Adjetivo Corto (1,2 Syllables) | -er | -est |
| Adjetivo Largo (3 or more) | more | most |




## 5.- Adverbios 

Los adverbios describen todo aquello que no sean sustantivos / nouns

### You can use two adverbs (one describing the other).

- She speaks English quite well.
- He drives really fast.
- They walk very slowly.



### Posiciones:

`Antes del verbo principal:` Los adverbios pueden colocarse antes del verbo principal en la oración. Por ejemplo:

- She quickly ran to catch the bus. (Ella corrió rápidamente para alcanzar el autobús)
- He often eats pizza for dinner. (Él a menudo come pizza para cenar)

`Después del verbo principal:` En algunas ocasiones, los adverbios se colocan después del verbo principal. Por ejemplo:

- She ran quickly to catch the bus.
- He eats pizza often for dinner.

`Al principio o al final de la oración:` Los adverbios también pueden aparecer al principio o al final de una oración para enfatizar una idea o para conectar la oración con el contexto anterior o posterior. Por ejemplo:

- Quickly, she ran to catch the bus.
- He eats pizza for dinner often.

`Antes del adjetivo o de otro adverbio:` Los adverbios pueden modificar tanto adjetivos como otros adverbios, y a menudo se colocan antes de ellos. Por ejemplo:

- She is very quick. (Ella es muy rápida)
- He speaks English quite fluently. (Él habla inglés bastante fluido)

`Los adverbios van siempre alrededor de un verbo, o al inicio o final de una oracion`

`Si una palabra con -ly viene antes de un sustantivo (noun) , eso sería el error.`

Tipo de adverbios : https://www.britishcouncil.org.mx/blog/adverbios-ingles#:~:text=Happy%20(feliz)%20%7C%20Happily%20(felizmente)
